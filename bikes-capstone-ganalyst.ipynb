## About the data:
ride_legnth is in seconds.


I'll turn all to char, then numeric ride length and after that datetime to start and end columns

```r
all_trips<- lapply(all_trips,as.character)
all_trips$ride_length<-as.numeric(all_trips$ride_length)
all_trips$started_at<-as_datetime(all_trips$started_at)
all_trips$ended_at<-as_datetime(all_trips$ended_at)
#str(all_trips)
```

```r
skim_without_charts(all_trips)
#although not all data is full, yet it is still useable.
```

```r
AllTrips<-as.data.frame(all_trips)
```

```r
AllTrips%>% ggplot(aes(x=ride_length))+geom_histogram()
#Clearly the majority of rides are below 4 minutes
```
